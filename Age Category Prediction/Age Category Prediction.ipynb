```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
data=pd.read_csv('Age Category.csv')
```

```python
data
```

```python
data['Class'].replace(['YOUNG','MIDDLE','OLD'],[0,1,2],inplace=True)
```

```python
data
```

```python
def read_img(path,ch=3,resize=(150,150)):
    di=tf.io.read_file(path)
    di=tf.image.decode_jpeg(di,channels=ch)
    di=tf.image.convert_image_dtype(di,dtype=tf.float32)
    di=tf.image.resize(di,resize)
    return di
```

```python
def load_data(image_path,label):
    img=read_img(image_path,3,(150,150))
    return(img,label)
```

```python
PATH='Train'
image_paths=[]
for path in os.listdir(PATH):
    image_paths.append(PATH+'/'+path)
response_list=[]
for i in image_paths:
    _,tail=os.path.split(i)
    response=data.loc[data['ID']==tail]['Class'].values[0]
    response_list.append(response)
```

```python
image_paths
```

```python
response_list
```

```python
train_size=int(0.9*(len(image_paths)))
test_size=int(0.1*(len(image_paths)))
```

```python
train_size
```

```python
test_size
```

```python
train_set=tf.data.Dataset.from_tensor_slices((image_paths[:train_size],
                                             response_list[:train_size]))
test_set=tf.data.Dataset.from_tensor_slices((image_paths[train_size:],
                                            response_list[train_size:]))
```

```python
len(train_set)
```

```python
len(test_set)
```

```python
train_set=(train_set
          .map(load_data,num_parallel_calls=tf.data.AUTOTUNE)
          .batch(64)
          .prefetch(tf.data.AUTOTUNE))

test_set=(test_set
         .map(load_data,num_parallel_calls=tf.data.AUTOTUNE)
         .batch(64)
         .prefetch(tf.data.AUTOTUNE))
```

```python
list(train_set)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
cnn=Sequential([
    Conv2D(filters=30,kernel_size=(3,3),activation='relu',input_shape=(150,150,3)),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(64,activation='relu'),
    Dense(3,activation='sigmoid')
])
```

```python
cnn.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.summary()
```

```python
cnn.fit(train_set,epochs=10,validation_data=test_set)
```

```python
test_loss,test_accuracy=cnn.evaluate(test_set)
```

```python
from tensorflow.keras.preprocessing.image import img_to_array,array_to_img,load_img,ImageDataGenerator
```

```python
def predict_image(img_path):
    img=load_img(img_path,target_size=(150,150))
    img_array=img_to_array(img)
    img_array=np.expand_dims(img_array,axis=0)
    img_array /=255.0
    
    prediction=cnn.predict(img_array)
    class_index=np.argmax(prediction,axis=1)
    
    class_label=['YOUNG','MIDDLE','OLD']
    prediction_class=class_label[class_index[0]]
    print(f'Predicted Class:{prediction_class}')
```
